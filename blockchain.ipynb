```python
import time
import hashlib

start_time = time.time()

import hashlib
import base58

# convert decimal number to hex string
hex_string = hex(7589)[2:]

# prepend with 00
hex_string = '00' + hex_string

# convert to binary
binary_data = bytes.fromhex(hex_string)

# compute SHA-256 hash
sha256_1 = hashlib.sha256(binary_data).digest()

# compute RIPEMD-160 hash
ripemd160 = hashlib.new('ripemd160')
ripemd160.update(sha256_1)

# prepend with network byte (0x00 for Mainnet, 0x6f for Testnet)
network_byte = b'\x00'
binary_data = network_byte + ripemd160.digest()

# compute double SHA-256 hash
sha256_2 = hashlib.sha256(hashlib.sha256(binary_data).digest()).digest()

# take first 4 bytes as checksum
checksum = sha256_2[:4]

# append checksum to binary data
binary_data += checksum

# encode in Base58
bitcoin_address = base58.b58encode(binary_data)

#print(bitcoin_address)


end_time = time.time()

print("Bitcoin public key hash: ", bitcoin_address)
print("Time taken: ", end_time - start_time, "seconds")
```

```python
import hashlib
import datetime
import time

start_time = time.time()
difficulty = 0
class Block:
    def __init__(self, data, previous_block_hash, difficulty):
        self.timestamp = datetime.datetime.now()
        self.data = data
        self.previous_block_hash = previous_block_hash
        self.nonce = 0
        self.hash = self.calculate_hash(difficulty)

    def calculate_hash(self, difficulty):
        sha = hashlib.sha256()
        hash_str = str(self.timestamp) + str(self.data) + str(self.previous_block_hash) + str(self.nonce)
        sha.update(hash_str.encode('utf-8'))
        hashed = sha.hexdigest()

        # Add difficulty level
        prefix = "0" * difficulty
        while not hashed.startswith(prefix):
            self.nonce += 1
            hash_str = str(self.timestamp) + str(self.data) + str(self.previous_block_hash) + str(self.nonce)
            sha = hashlib.sha256()
            sha.update(hash_str.encode('utf-8'))
            hashed = sha.hexdigest()

        return hashed

class Blockchain:
    def __init__(self, difficulty):
        self.chain = [self.create_genesis_block()]
        self.difficulty = difficulty

    def create_genesis_block(self):
        return Block("Genesis Block", "0", difficulty)

    def add_block(self, new_block):
        new_block.previous_block_hash = self.chain[-1].hash
        new_block.hash = new_block.calculate_hash(self.difficulty)
        self.chain.append(new_block)

    def validate_chain(self):
        for i in range(1, len(self.chain)):
            current_block = self.chain[i]
            previous_block = self.chain[i-1]

            if current_block.hash != current_block.calculate_hash(self.difficulty):
                return False

            if current_block.previous_block_hash != previous_block.hash:
                return False

        return True

if __name__ == '__main__':
    
    blockchain = Blockchain(3) # Set difficulty level 

    block1 = Block("Block 1 Data", "", blockchain.difficulty)
    blockchain.add_block(block1)

    block2 = Block("Block 2 Data", "", blockchain.difficulty)
    blockchain.add_block(block2)

    '''block3 = Block("Block 3 Data", "", blockchain.difficulty)
    blockchain.add_block(block3)

    block4 = Block("Block 4 Data", "", blockchain.difficulty)
    blockchain.add_block(block4)

    block5 = Block("Block 5 Data", "", blockchain.difficulty)
    blockchain.add_block(block5)

    block6 = Block("Block 6 Data", "", blockchain.difficulty)
    blockchain.add_block(block6)'''

    # Validate the chain
    print(f"Is the blockchain valid? {blockchain.validate_chain()}")
end_time = time.time()

print(end_time-start_time)
```
